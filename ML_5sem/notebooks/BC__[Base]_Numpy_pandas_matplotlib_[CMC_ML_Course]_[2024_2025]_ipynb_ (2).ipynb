Основные типы графиков:
- <font color='DarkSeaGreen'>**line plot**</font> — обычный поточечный график, которым можно изображать кривые, соединяющие последовательные точки;
- <font color='DarkSeaGreen'>**histogram**</font> — гистограмма, показывающая распределение некоторое величины;
- <font color='DarkSeaGreen'>**scatter plot**</font> — диаграмма рассеивания, показывающая взаимосвязь двух величин;
- <font color='DarkSeaGreen'>**bar plot**</font> — столбчатая диаграмма, показывающий взаимосвязь количественной величины от категориальной.

**Возьмем логарифм от данных для наглядности:**

```python
import numpy as np

log_view_count = np.log10(france_videos.view_count + 1)
log_comment_count = np.log10(france_videos.comment_count + 1)

plt.figure(figsize=(12, 7))
plt.scatter(x=log_view_count, y=log_comment_count, color='blue')
plt.title("Логарифмический график зависимости кол-ва комментариев от кол-ва просмотров для роликов во Франции.")
plt.xlabel('Кол-во просмотров')
plt.ylabel('Кол-во комментариев')
plt.grid()

plt.tight_layout()
plt.show()
```

В то же время есть единичные случаи, когда просмотров было больше __(1e7)__, но комментариев гораздо меньше __(175 тысяч)__.

На логарифмическом графике зависимости тоже наблюдается, что несмотря на просмотр ролика зрители могли и не оставлять комментарии.

Поскольку кол-во просмотров растет, а кол-во комментарев может не изменяться, то характер зависимости - **__нелинейный__**.
