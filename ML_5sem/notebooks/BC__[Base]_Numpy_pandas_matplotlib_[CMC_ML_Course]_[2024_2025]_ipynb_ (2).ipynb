```python
df.publishedAt = pd.to_datetime(df.publishedAt)
```

```python
df.publishedAt.dt.date
```

```python
import matplotlib.pyplot as plt #     

# your code here

unique_videos = df[['video_id', 'publishedAt']].drop_duplicates()
video_counts = unique_videos.groupby(unique_videos['publishedAt'].dt.date).count()

plt.figure(figsize=(12, 7))
plt.plot(video_counts.index, video_counts['video_id'], marker='o')
plt.title("  -     ")
plt.xlabel(' ')
plt.ylabel(' ')
plt.xticks(rotation=45)
plt.grid()

plt.tight_layout()
plt.show()
```

### <font color='DarkOrange'>** 2 [0.5 a]**</font>

  `(bar plot)` -      .

```python
# your code here

unique_videos_by_country = df[['video_id', 'country']].drop_duplicates()
video_counts_by_country = unique_videos_by_country['country'].value_counts()

plt.figure(figsize=(12, 7))
video_counts_by_country.plot(kind='bar', color='skyblue')
plt.title(" -      .")
plt.xlabel('')
plt.ylabel('-  ')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

### <font color='DarkOrange'>** 3 [1 ]**</font>

  `(scatter plot)`  -     -      (country='FR').          (comments_disabled=True).

    ?

```python
# your code here

france_videos = df[(df['country'] == 'FR') & (df['comments_disabled'] == False)]

plt.figure(figsize=(12, 7))
plt.scatter(x=france_videos['view_count'], y=france_videos['comment_count'], color='skyblue')
plt.title("  -     -     .")
plt.xlabel('- ')
plt.ylabel('- ')
plt.grid()

plt.tight_layout()
plt.show()
```

**     :**

```python
import numpy as np

log_view_count = np.log10(france_videos.view_count + 1)
log_comment_count = np.log10(france_videos.comment_count + 1)

plt.figure(figsize=(12, 7))
plt.scatter(x=log_view_count, y=log_comment_count, color='blue')
plt.title("   -   -     .")
plt.xlabel('- ')
plt.ylabel('- ')
plt.grid()

plt.tight_layout()
plt.show()
```
