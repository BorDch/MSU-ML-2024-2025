```python
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets

def plot_knn_bound(X, y, scaler=None, n_neighbors=10, xlim=(-15, 15), ylim=(-20, 20)):
    # step size in the mesh
    h = 0.05

    # Create color maps
    cmap_light = ListedColormap(['C0', 'orange', 'cyan', 'green'][:np.unique(y).shape[0]])
    cmap_bold = ['C0', 'orange', 'c', 'darkgreen'][:np.unique(y).shape[0]]

    x_min, x_max = xlim
    y_min, y_max = ylim
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    grid = np.c_[xx.ravel(), yy.ravel()]

    X_scaled = X # if scaler is None
    if scaler is not None:
        grid = scaler.transform(grid)
        X_scaled = scaler.transform(X)

    # we create an instance of Neighbours Classifier and fit the data.
    clf = neighbors.KNeighborsClassifier(n_neighbors, algorithm='brute')
    clf.fit(X_scaled, y)

    Z = clf.predict(grid)

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y,
                    palette=cmap_bold, alpha=1.0, edgecolor="black")
    plt.xlabel('x_1')
    plt.ylabel('x_2')
    plt.title('   {}-NN'.format(n_neighbors))
    plt.grid()
    plt.show()
```

   $1$-$NN$  $10$-$NN$

```python
plot_knn_bound(X, y, n_neighbors=1)
plot_knn_bound(X, y, n_neighbors=10)
```

```python
from sklearn import neighbors

#    
best_params = max(res, key = lambda x: res[x])
print(best_params, res[best_params]) # ,   ('TfidfVectorizer', 1, 'cosine', 'uniform')

#     
X_train_vec = tf_idf.fit_transform(X_train)
X_test_vec = tf_idf.transform(X_test)

#     
model = neighbors.KNeighborsClassifier(1, weights = 'uniform', metric = 'cosine')
model.fit(X_train_vec, y_train)
y_pred = model.predict(X_test_vec)
accuracy_score(y_test, y_pred)
```

   : ('TfidfVectorizer', 1, 'cosine', 'uniform')

**      **

```python
#   ,   

# 
data_id = np.arange(20)
data_train = []

for i in range(len(data_id)):
  y_pred = model.predict(X_train_vec[i == y_train])
  data_train.append(accuracy_score(y_train[i == y_train], y_pred))

# 
data_test = []

for i in range(len(data_id)):
  y_pred = model.predict(X_test_vec[i == y_test])
  data_test.append(accuracy_score(y_test[i == y_test], y_pred))
```

```python
# 

plt.figure(figsize=(12, 6))
plt.plot(data_id, data_train, marker='o')
plt.plot(data_id, data_test, marker='o')
plt.title('     ')
plt.xlabel(' ()')
plt.ylabel(' ')
plt.xticks(data_id)
plt.legend([' ', ' '], loc='center left')
plt.grid()
plt.show()
```
